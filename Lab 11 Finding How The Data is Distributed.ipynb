### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
## Write your code here
df.describe()
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Write your code here
def fill_all_modes(df):
    for col in df.columns:
        if df[col].isnull().any():
            mode_val = df[col].mode()[0]
            df[col] = df[col].fillna(mode_val)
            print(f'Colonne "{col}" imputÃ©e avec "{mode_val}". Valeurs manquantes restantes : {df[col].isna().sum()}')
            
fill_all_modes(df)

df_clean = df
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
df['Employment'] = df['Employment'].astype(str).fillna('')

df_exploded_emp = df.assign(EmpList=df['Employment'].str.split(';')).explode('EmpList')

df_exploded_emp['EmpList'] = df_exploded_emp['EmpList'].str.strip()

df_exploded_emp = df_exploded_emp[df_exploded_emp['EmpList'] != '']
df_exploded_emp = df_exploded_emp[df_exploded_emp['EmpList'].notna()]

df_clean = df_exploded_emp
```

```python
## Write your code here
df_clean['YearsCodePro'] = df_clean['YearsCodePro'].astype(str)
df_clean['YearsCodePro'] = df_clean['YearsCodePro'].replace({
    'Less than 1 year': '1',
    'More than 50 years': '50'
})

print(df_clean['EmpList'].value_counts(ascending=False))
print(df_clean['JobSat'].value_counts(ascending=False))
print(df_clean['YearsCodePro'].value_counts(ascending=False))
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here
df['JobSat'].value_counts().plot(kind = 'pie', autopct = '%1.0f%%', figsize = (6,6), shadow = True, pctdistance = 0.50, labels = None)
plt.title('Repartion of Job Satisfaction')
plt.legend(df['JobSat'].value_counts().index, loc = 'upper left')
plt.axis('equal')
plt.show()
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here
# 1. Exploser les deux colonnes
df_hw = df_clean['LanguageHaveWorkedWith'].str.split(';').explode().str.strip()
df_wtw = df_clean['LanguageWantToWorkWith'].str.split(';').explode().str.strip()

# 2. Calculer les frÃ©quences
freq_hw = df_hw.value_counts()
freq_wtw = df_wtw.value_counts()

# 3. Combiner en un seul DataFrame
df_lang = pd.concat([freq_hw, freq_wtw], axis=1, keys=['HaveWorkedWith', 'WantToWorkWith']).fillna(0)

# 4. top 10 langages (basÃ© sur les 2 colonnes cumulÃ©es)
top_langs = df_lang.sum(axis=1).sort_values(ascending=False).head(10).index
df_top = df_lang.loc[top_langs]

# 5. Graphique
fig, ax = plt.subplots(figsize=(10, 6))
df_top.plot(kind='bar', stacked=True, ax=ax)
ax.set_xlabel('Programming Language')
ax.set_ylabel('Frequency')
ax.set_title('Comparison of Programming Languages: Used vs Wanted')
ax.legend()
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{int(height)}', 
                (p.get_x() + p.get_width() / 2., height), 
                ha='center', va='bottom', fontsize=8)
plt.title("Comparison of Programming Languages: Used vs Wanted")
plt.ylabel("Frequency")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here
df_grp_RW_Country = df_clean.groupby('Country')['RemoteWork'].value_counts(ascending=False)

df_pivot_RW_Country = df_grp_RW_Country.unstack(fill_value=0)

plt.figure(figsize = (15,20))
sns.heatmap(df_pivot_RW_Country, fmt = 'd', cmap = 'YlGnBu')
plt.xlabel('RemoteWork')
plt.ylabel('Country')
plt.title('RemoteWork by Country')
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
## Write your code here

df_clean['YearsCodePro'] = pd.to_numeric(df_clean['YearsCodePro'], errors='coerce')

correlation_JS_YC = df_clean['JobSat'].corr(df_clean['YearsCodePro'])
print(f'La corrÃ©lation entre JobSat et YearsCodePro est de : {correlation_JS_YC:.3f}')

pearson_corr = df_clean[['YearsCodePro', 'JobSat']].corr(method='pearson')

pearson_value = pearson_corr.loc['YearsCodePro', 'JobSat']
print(f'La corrÃ©lation de Pearson entre JobSat et YearsCodePro est de : {pearson_value:.3f}')

spearman_corr = df_clean[['YearsCodePro', 'JobSat']].corr(method='spearman')

spearman_value = spearman_corr.loc['YearsCodePro', 'JobSat']
print(f'La corrÃ©lation de Spearman entre JobSat et YearsCodePro est de : {spearman_value:.3f}')

sns.regplot(x = 'YearsCodePro', y = 'JobSat', data = df_clean, line_kws = {'color' : 'red'})
plt.title('Correlation entre JobSat et YearsCodePro')
plt.xlabel('YearsCodePro')
plt.ylabel('JobSat')
plt.show()
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here
crosstab_Ed_Emp = pd.crosstab(df_exploded_emp['EmpList'], df_exploded_emp['EdLevel'])
crosstab_Ed_Emp.plot(kind = 'bar', figsize=(12,8), stacked = True)
plt.title('RÃ©partition des niveaux d\'Ã©tudes par emplois')
plt.xticks(rotation = 45)
plt.xlabel('Emplois')
plt.ylabel('Nombre de personnes')
plt.legend(title='Niveau d\'Ã©tude')
```
