### Task 1: Exploring Data Distributions Using Bubble Plots


#### 1. Bubble Plot for Age vs. Frequency of Participation


- Visualize the relationship between respondents’ age and their participation frequency (`SOPartFreq`) using a bubble plot.

- Use the size of the bubbles to represent their job satisfaction (`JobSat`).


```python
##Write your code here
df1 = df[['Age', 'SOPartFreq', 'JobSat']]
df1.dropna(subset = ['Age', 'SOPartFreq', 'JobSat'], inplace = True)
df1 = df1[df1['Age'] != 'Prefer not to say']

age_order = [
    'Under 18 years old', '18-24 years old', '25-34 years old',
    '35-44 years old', '45-54 years old', '55-64 years old',
    '65 years or older']
    
df1['Age'] = pd.Categorical(df1['Age'], categories=age_order, ordered=True)

df1_grp = df1.groupby(['Age', 'SOPartFreq'])['JobSat'].mean()

plt.scatter(x=df1_grp.index.get_level_values('Age'),
            y=df1_grp.index.get_level_values('SOPartFreq'),
            s=df1_grp.values * 40,
            color='blue')

plt.xlabel('Age')
plt.ylabel('SOPartFreq')
plt.xticks(rotation = 45)
plt.show()
```

#### 2. Bubble Plot for Compensation vs. Job Satisfaction


-Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSat`).

- Use the size of the bubbles to represent respondents’ age.


```python
##Write your code here

df2 = df[['ConvertedCompYearly', 'JobSat', 'Age']]
df2.dropna(subset = ['ConvertedCompYearly', 'JobSat', 'Age'], inplace = True)
df2 = df2[df2['Age'] != 'Prefer not to say']

df2['ConvertedCompYearly'] = pd.to_numeric(df2['ConvertedCompYearly'], errors='coerce').astype('float')



age_order = [
    'Under 18 years old', '18-24 years old', '25-34 years old',
    '35-44 years old', '45-54 years old', '55-64 years old', '65 years or older']

df2['Age'] = pd.Categorical(df2['Age'], categories=age_order, ordered=True)

age_counts_df = df2['Age'].value_counts().reset_index()
age_counts_df.columns = ['Age', 'Age_counts']

df2_grp = df2.groupby('Age')[['ConvertedCompYearly','JobSat']].mean().round(2)

new_df = pd.merge(df2_grp, age_counts_df, on='Age')

age_size_map = {age: (i + 1) * 150 for i, age in enumerate(age_order)}
age_color_map = {age: color for age, color in zip(age_order, ['red', 'orange', 'gold', 'green', 'blue', 'purple', 'gray'])}

new_df['AgeSize'] = new_df['Age'].map(age_size_map)
new_df['AgeColor'] = new_df['Age'].map(age_color_map)

plt.figure(figsize=(12, 6))
plt.scatter(
    x=new_df['ConvertedCompYearly'],
    y=new_df['JobSat'],
    s=new_df['AgeSize'],
    c=new_df['AgeColor'],
    alpha=0.6,
    edgecolors='k')

for age in age_order:
    plt.scatter([], [], c=age_color_map[age], s=100, label=age)

plt.legend(title='Tranche d’âge', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title('Bubble Plot for Compensation vs. Job Satisfaction')
plt.xlabel('ConvertedCompYearly')
plt.ylabel('JobSat')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 2: Analyzing Relationships Using Bubble Plots


#### 1. Bubble Plot of Technology Preferences by Age

- Visualize the popularity of programming languages respondents have worked with (`LanguageHaveWorkedWith`) across age groups.

- Use bubble size to represent the frequency of each language.



```python
##Write your code here
df3 = df[['Age', 'LanguageHaveWorkedWith']]
df3.dropna(subset = ['Age', 'LanguageHaveWorkedWith'], inplace = True)
df3 = df3[df3['Age'] != 'Prefer not to say']

age_order = [
    'Under 18 years old', '18-24 years old', '25-34 years old',
    '35-44 years old', '45-54 years old', '55-64 years old', '65 years or older']

df3['Age'] = pd.Categorical(df3['Age'], categories=age_order, ordered=True)

df3['LanguageList'] = df3['LanguageHaveWorkedWith'].str.split(';').explode('LanguageList')
df3['LanguageList'] = df3['LanguageList'].str.strip()

df3.drop(columns=['LanguageHaveWorkedWith'], inplace=True)


language_counts_df = df3['LanguageList'].value_counts().reset_index()
language_counts_df.columns = ['LanguageList', 'count'] 
df3_mrg = df3.merge(language_counts_df, on='LanguageList').head(10)

plt.figure(figsize=(15, 8))
plt.scatter(x=df3_mrg['Age'], y=df3_mrg['LanguageList'], s=df3_mrg['count'], c = df3_mrg['count'], cmap = 'rainbow', alpha=0.6, edgecolors='k') 
plt.title('Bubble Plot of Top 10 Technology Preferences by Age')
plt.xlabel('Age')
plt.ylabel('Programming Language')
plt.colorbar(label='count')
plt.xticks(rotation = 45)
plt.grid(True)
plt.tight_layout() 
plt.show()
```

#### 2. Bubble Plot for Preferred Databases vs. Job Satisfaction

- Explore the relationship between preferred databases (`DatabaseWantToWorkWith`) and job satisfaction.

- Use bubble size to indicate the number of respondents for each database.


```python
##Write your code here

df4 = df[['DatabaseWantToWorkWith', 'JobSat']]
df4.dropna(subset = ['DatabaseWantToWorkWith', 'JobSat'], inplace = True)

df4['DatabaseList'] = df4['DatabaseWantToWorkWith'].str.split(';').explode('DatabaseList')
df4['DatabaseList'] = df4['DatabaseList'].str.strip()

df4.drop(columns=['DatabaseWantToWorkWith'], inplace=True)

data_counts_df = df4['DatabaseList'].value_counts().reset_index()
data_counts_df.columns = ['DatabaseList', 'Numbers of Respondants']

df4_grp = df4.groupby('DatabaseList')['JobSat'].mean().round(2).sort_values(ascending=False)

df4_mrg = pd.merge(df4_grp, data_counts_df, on='DatabaseList').head(10)

plt.figure(figsize=(15, 8))
plt.scatter(x=df4_mrg['DatabaseList'], y=df4_mrg['JobSat'], s=df4_mrg['Numbers of Respondants']*2, c=df4_mrg['Numbers of Respondants'], cmap = 'rainbow', alpha=0.6, edgecolors='k')
plt.title('Bubble Plot for Top 10 Preferred Databases vs. Job Satisfaction')
plt.xlabel('DatabaseList')
plt.ylabel('JobSat')
plt.colorbar(label='Numbers of Respondants')
plt.xticks(rotation = 45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 3: Comparing Data Using Bubble Plots


#### 1. Bubble Plot for Compensation Across Developer Roles

- Visualize compensation (`ConvertedCompYearly`) across different developer roles (`DevType`).

- Use bubble size to represent job satisfaction.


```python
##Write your code here

df5 = df[['DevType', 'ConvertedCompYearly', 'JobSat']]

df5.dropna(subset = ['DevType', 'ConvertedCompYearly', 'JobSat'], inplace = True)

df5['ConvertedCompYearly'] = pd.to_numeric(df5['ConvertedCompYearly'], errors='coerce').astype('float')

df5_grp = df5.groupby('DevType')[['ConvertedCompYearly', 'JobSat']].median().round(2)

plt.figure(figsize=(15,10))
plt.scatter(x=df5_grp['ConvertedCompYearly'], y=df5_grp.index, s=df5_grp['JobSat']*100, c=df5_grp['JobSat'], cmap = 'rainbow', alpha=0.6, edgecolors='k')
plt.title('Bubble Plot for Compensation Across Developer Roles')
plt.xlabel('ConvertedCompYearly')
plt.ylabel('DevType')
plt.colorbar(label='JobSat')
plt.xticks(rotation = 45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Collaboration Tools by Age

- Visualize the relationship between the collaboration tools used (`NEWCollabToolsHaveWorkedWith`) and age groups.

- Use bubble size to represent the frequency of tool usage.


```python
##Write your code here

df6 = df[['Age', 'NEWCollabToolsHaveWorkedWith']]
df6.dropna(subset = ['Age', 'NEWCollabToolsHaveWorkedWith'], inplace = True)
df6 = df6[df6['Age'] != 'Prefer not to say']

age_order = [
    'Under 18 years old', '18-24 years old', '25-34 years old',
    '35-44 years old', '45-54 years old', '55-64 years old',
    '65 years or older']
df6['Age'] = pd.Categorical(df6['Age'], categories=age_order, ordered=True)

df6['ToolsList'] = df6['NEWCollabToolsHaveWorkedWith'].str.split(';').explode('ToolsList')
df6['ToolsList'] = df6['ToolsList'].str.strip()

df6.drop(columns=['NEWCollabToolsHaveWorkedWith'], inplace=True)

tools_counts_df = df6['ToolsList'].value_counts().reset_index()
tools_counts_df.columns = ['ToolsList', 'Tools_counts']

top10_tools = tools_counts_df.head(10)

df6_mrg = df6.merge(top10_tools, on='ToolsList')

df6_grp = df6_mrg.groupby(['Age', 'ToolsList'])['Tools_counts'].count()

plt.figure(figsize=(20, 8))
plt.scatter(x= df6_grp.index.get_level_values('Age'),
            y = df6_grp.index.get_level_values('ToolsList'),
            s = df6_grp.values, 
            c = df6_grp.values, 
            cmap = 'rainbow', alpha=0.6, edgecolors='k')
plt.title('Bubble Plot for Collaboration Tools by Age')
plt.xlabel('Age')
plt.ylabel('Collaboration Tools')
plt.colorbar(label='Tools_counts')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

### Task 4: Visualizing Technology Trends Using Bubble Plots


#### 1. Bubble Plot for Preferred Web Frameworks vs. Job Satisfaction

- Explore the relationship between preferred web frameworks (`WebframeWantToWorkWith`) and job satisfaction.

- Use bubble size to represent the number of respondents.



```python
##Write your code here

df7 = df[['WebframeWantToWorkWith','JobSat']]
df7.dropna(subset = ['WebframeWantToWorkWith', 'JobSat'], inplace = True)

df7['FrameList'] = df7['WebframeWantToWorkWith'].str.split(';').explode('FrameList')
df7['FrameList'] = df7['FrameList'].str.strip()

df7.drop(columns=['WebframeWantToWorkWith'], inplace=True)

frame_counts_df = df7['FrameList'].value_counts().reset_index()
frame_counts_df.columns = ['FrameList', 'Numbers of Respondants']

df7_mrg = df7.merge(frame_counts_df, on='FrameList')

df7_grp = df7_mrg.groupby(['FrameList','Numbers of Respondants'])['JobSat'].mean().round(2).sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 4))
plt.scatter(x = df7_grp.index.get_level_values('FrameList'), 
            y = df7_grp.values,                             
            s = df7_grp.index.get_level_values('Numbers of Respondants')*2, 
            c = df7_grp.index.get_level_values('Numbers of Respondants'), 
            cmap = 'rainbow',
            alpha=0.6,
            edgecolors='k')
plt.title('Top 10 Bubble Plot for Preferred Web Frameworks vs. Job Satisfaction')
plt.xlabel('FrameList')
plt.ylabel('JobSat') # Y-axis is JobSat, not Numbers of Respondants
plt.colorbar(label='Numbers of Respondants')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```
