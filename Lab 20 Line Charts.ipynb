### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
## Write your code here

df1 = df[['Age','ConvertedCompYearly']].dropna()
df1 = df1[df1['Age'] != 'Prefet not to say']
df1['ConvertedCompYearly'] = pd.to_numeric(df1['ConvertedCompYearly'], errors='coerce')

age_order = [
    'Under 18 years old', '18-24 years old', '25-34 years old',
    '35-44 years old', '45-54 years old', '55-64 years old', '65 years or older']

df1 = df1.groupby('Age')['ConvertedCompYearly'].mean().reset_index().sort_values(by='ConvertedCompYearly', ascending = False).astype({'ConvertedCompYearly': float}).round(2)

df1['Age'] = pd.Categorical(df1['Age'], categories=age_order, ordered=True)

df1 = df1.sort_values('Age').dropna()

plt.figure(figsize = (10, 6))
df1.plot(kind = 'line', x = 'Age', y = 'ConvertedCompYearly', marker = 'o', color = 'blue')
plt.xlabel('Age')
plt.ylabel('Converted Compensation Yearly')
plt.title('Converted Compensation Yearly by Age Group')
plt.xticks(rotation = 40)
plt.grid(True)
plt.tight_layout()
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
## Write your code here
df2 = df1.copy()

df2 = df2[df2['Age'].isin(['25-34 years old', '35-44 years old', '45-54 years old'])]

age_mapping = {
    '25-34 years old': 25,
    '35-44 years old': 35,
    '45-54 years old': 45 }

df2['Age_mean'] = df2['Age'].map(age_mapping)

plt.figure(figsize = (10, 6))
df2.plot(kind = 'line', x = 'Age_mean', y = 'ConvertedCompYearly', marker = 'o', color = 'magenta')
plt.xlabel('Age')
plt.ylabel('Converted Compensation Yearly')
plt.title('Converted Compensation Yearly by Age Group')
plt.xticks(rotation = 45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
## Write your code here
df3 = df[['WorkExp','JobSatPoints_6']].dropna()

df3 = df3.groupby('WorkExp')['JobSatPoints_6'].mean().reset_index().round(2)

df3.plot(kind = 'line', x = 'WorkExp', y = 'JobSatPoints_6', marker ='o', color = 'green', figsize=(12,8))
plt.xlabel('Working Experience') 
plt.ylabel('Mean of JobSatPoints_6')
plt.title('Mean of JobSatPoints_6 by Working Experience') 
plt.legend('')
plt.grid(True) 
plt.tight_layout() 
plt.show() 
```

### Task 3: Trends in Job Satisfaction and Compensation by Experience


##### 1.Line Chart of Median ConvertedCompYearly Over Experience Level

- This line chart will track how median compensation (`ConvertedCompYearly`) changes with increasing experience.

- Use a column such as `WorkExp` or another relevant experience-related column.


```python
## Write your code here

df4 = df[['ConvertedCompYearly', 'WorkExp']].dropna()

df4 = df4.groupby('WorkExp')['ConvertedCompYearly'].median().reset_index().round(2)

df4.plot(kind = 'line', x = 'WorkExp', y = 'ConvertedCompYearly', marker = 'o', color = 'purple', figsize = (12,8)) 
plt.xlabel('Working Experience') 
plt.ylabel('Median of Yearly Compensation') 
plt.grid(True) 
plt.legend('') 
plt.title('Median of Yearly Compensation by Working Experience')
plt.show()
```

##### 2.Line Chart of Job Satisfaction (`JobSatPoints_6`) Across Experience Levels

- Create a line chart to explore trends in job satisfaction (`JobSatPoints_6`) based on experience level.

- This chart will provide insight into how satisfaction correlates with experience over time


```python
## Write your code here

df5 = df[['WorkExp','JobSatPoints_6']].dropna()

df5 = df5.groupby('JobSatPoints_6')['WorkExp'].mean().reset_index().round(2)

df5.plot(kind = 'line', x = 'JobSatPoints_6', y = 'WorkExp', marker ='o', color = 'green', figsize=(12,8))
plt.xlabel('Working Experience') 
plt.ylabel('Mean of Working Experience')
plt.title('Mean of Working Experience by JobSatPoints_6') 
plt.legend('')
plt.grid(True) 
plt.tight_layout() 
plt.show() 
```
