**Download and Load the Data**

### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here
df1 = df['DatabaseWantToWorkWith'].dropna().to_frame()

df1['DataList'] = df1['DatabaseWantToWorkWith'].str.split(';').explode('DataList')
df1['DataList'] = df1['DataList'].str.strip()

data_counts = df1['DataList'].value_counts().sort_values(ascending=False).head(5)

data_counts.plot(kind = 'pie', autopct = '%1.1f%%')
plt.title('Top 5 Databases')
plt.ylabel('')
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
##Write your code here
df2 = df['DevType'].dropna().to_frame()

dev_counts = df2['DevType'].value_counts().head(5)

plt.pie(dev_counts, labels=dev_counts.index, autopct='%1.1f%%')
plt.title('Top 5 Developer Roles')
plt.ylabel('')
plt.show()
```

##### 1.3 Create a pie chart for the operating systems used by respondents for professional use


The `OpSysProfessional` use column shows the operating systems developers use professionally. Let’s visualize the distribution of the top operating systems in a pie chart.



```python
##Write your code here
df3 = df['OpSysProfessional use'].dropna().to_frame()

df3['OpSysProfessional use'] = df3['OpSysProfessional use'].str.split(';').explode('OpSysProfessional use')
df3['OpSysProfessional use'] = df3['OpSysProfessional use'].str.strip()

op_counts = df3['OpSysProfessional use'].value_counts().sort_values(ascending = False).head(5)

plt.pie(op_counts, labels = op_counts.index, autopct = '%1.1f%%')
plt.title('Top 5 Operating Systems')
plt.ylabel('')
plt.show()
```

### Task 2: Additional Visualizations and Comparisons


##### 2.1 Pie Chart for Top 5 Programming Languages Respondents Have Worked With


The `LanguageHaveWorkedWith` column contains the programming languages that respondents have experience with. We’ll plot a pie chart to display the composition of the top 5 languages.



```python
##Write your code here

df4 = df['LanguageHaveWorkedWith'].dropna().to_frame()
df4['LanguageList'] = df4['LanguageHaveWorkedWith'].str.split(';').explode('LanguageList')
df4['LanguageList'] = df4['LanguageList'].str.strip()

Language_counts = df4['LanguageList'].value_counts().sort_values(ascending = False).head(5)

plt.pie(Language_counts, labels = Language_counts.index, autopct= '%1.1f%%')
plt.title('Top 5 Languages')
plt.ylabel('')
plt.show()
```

##### 2.2 Pie Chart for Top Collaboration Tools used in Professional Use


Using the `NEWCollabToolsHaveWorkedWith` column, we’ll identify and visualize the top collaboration tools respondents use in their professional work.



```python
##Write your code here

df5 = df['NEWCollabToolsHaveWorkedWith'].dropna().to_frame()

df5['ToolsList'] = df5['NEWCollabToolsHaveWorkedWith'].str.split(';').explode('ToolsList')
df5['ToolsList'] = df5['ToolsList'].str.strip()

Tools_counts = df5['ToolsList'].value_counts().sort_values(ascending = False).head(5)

plt.pie(Tools_counts, labels = Tools_counts.index, autopct = '%1.1f%%')
plt.title('Top 5 Tools Respondants haved work with')
plt.show()
```

### Task 3: Analyzing and Interpreting Composition


In this task, you will create additional pie charts to analyze specific aspects of the survey data. Use `pandas` and `matplotlib` to complete each task and interpret the findings.



##### 3.1 Pie Chart of `Respondents` Most Admired Programming Languages


The `LanguageAdmired` column lists the programming languages respondents admire most. Create a pie chart to visualize the top 5 admired languages.



```python
##Write your code here

df6 = df['LanguageAdmired'].dropna().to_frame()

df6['LanguageList'] = df6['LanguageAdmired'].str.split(';').explode('LanguageList')
df6['LanguageList'] = df6['LanguageList'].str.strip()

Language_counts = df6['LanguageList'].value_counts().sort_values(ascending = False).head(5)

plt.pie(Language_counts, labels = Language_counts.index, autopct = '%1.1f%%')
plt.title('Top 5 Language Admired')
plt.show()
```

##### 3.2 Pie Chart of Tools Used for AI Development


Using the `AIToolCurrently` Using column, create a pie chart to visualize the top 5 tools developers are currently using for AI development.



```python
##Write your code here

df7 = df['AIToolCurrently Using'].dropna().to_frame()

df7['AI_List'] = df7['AIToolCurrently Using'].str.split(';').explode('AI_List') 
df7['AI_List'] = df7['AI_List'].str.strip() 

AI_counts = df7['AI_List'].value_counts().sort_values(ascending = False).head(5) 

plt.pie(AI_counts, labels = AI_counts.index, autopct= '%1.1f%%') 
plt.title('Top 5 AI Tools') 
plt.show() 
```

##### 3.3 Pie Chart for Preferred Web Frameworks


The `WebframeWantToWorkWith` column includes web frameworks that respondents are interested in working with. Visualize the top 5 frameworks in a pie chart.



```python
##Write your code here

df8 = df['WebframeWantToWorkWith'].dropna().to_frame()
df8['FrameList'] = df8['WebframeWantToWorkWith'].str.split(';').explode('FrameList')
df8['FrameList'] = df8['FrameList'].str.strip()

df8 = df8.dropna(subset = ['FrameList'])

Frame_counts = df8['FrameList'].value_counts().sort_values(ascending = False).head(5)

plt.pie(Frame_counts, labels = Frame_counts.index, autopct = '%1.1f%%') 
plt.title('Top 5 Web Frame') 
plt.show()
```

##### 3.4 Pie Chart for Most Desired Embedded Technologies


Using the `EmbeddedWantToWorkWith` column, create a pie chart to show the top 5 most desired embedded technologies that respondents wish to work with.



```python
##Write your code here

df9 = df['EmbeddedWantToWorkWith'].dropna().to_frame() 
df9['EmbeddedList'] = df9['EmbeddedWantToWorkWith'].str.split(';').explode('EmbeddedList')
df9['EmbeddedList'] = df9['EmbeddedList'].str.strip() 

df9 = df9.dropna(subset = ['EmbeddedList'])

Embedded_counts = df9['EmbeddedList'].value_counts().sort_values(ascending = False).head(5)

plt.pie(Embedded_counts, labels = Embedded_counts.index, autopct = '%1.1f%%') 
plt.title('Top 5 Embedded') 
plt.show()
```
