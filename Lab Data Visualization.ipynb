**Read the CSV File into a Pandas DataFrame**

Load the Stack Overflow survey data into a Pandas DataFrame:

**Create a SQLite Database and Insert the Data**

Now, let's create a new SQLite database (`survey-data.sqlite`) and insert the data from the DataFrame into a table using the sqlite3 library:


```python
import sqlite3

# Create a connection to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Write the dataframe to the SQLite database
df.to_sql('main', conn, if_exists='replace', index=False)


# Close the connection
conn.close()
```

**Verify the Data in the SQLite Database**
Verify that the data has been correctly inserted into the SQLite database by running a simple query:


```python
# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run a simple query to check the data
QUERY = "SELECT * FROM main LIMIT 5"
df_check = pd.read_sql_query(QUERY, conn)

# Display the results
print(df_check)
```

## Hands-on Lab


### Visualizing the Distribution of Data

**Histograms**

Plot a histogram of CompTotal (Total Compensation).


```python
CompTotal = """
SELECT DISTINCT ResponseId,

COALESCE(CompTotal, (SELECT ROUND(AVG(CompTotal),3) FROM main WHERE CompTotal IS NOT NULL)) AS Clean_CT

FROM main
"""
CT = pd.read_sql_query(CompTotal, conn)

CT['Clean_CT'] = pd.to_numeric(CT['Clean_CT'])

bins = [0, 20000, 40000, 60000, 80000, 100000, 150000, 200000, 500000, 1e6]
labels = ['0-20k', '20k-40k', '40k-60k', '60k-80k', '80k-100k',
          '100k-150k', '150k-200k', '200k-500k', '500k+']

CT['CompTotalRange'] = pd.cut(CT['Clean_CT'], bins=bins, labels=labels)

CT_counts = CT['CompTotalRange'].value_counts().sort_index()

fig, ax = plt.subplots(figsize=(8, 6))
plt.bar(CT_counts.index, CT_counts.values)
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{int(height)}',
                (p.get_x() + p.get_width() / 2., height),
                ha='center', va='bottom', fontsize=8, fontweight = 'bold')
plt.title('Compensation Range')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Box Plots**

Plot a box plot of Age.


```python
## Write your code here
Age = """
SELECT Age, ConvertedCompYearly
FROM main
WHERE ConvertedCompYearly IS NOT NULL AND Age IS NOT NULL
"""
Age = pd.read_sql_query(Age, conn)

Age = pd.DataFrame(Age)
Age['ConvertedCompYearly'] = pd.to_numeric(Age['ConvertedCompYearly'])
plt.figure(figsize=(15,8))
sns.boxplot(x = 'Age', y = 'ConvertedCompYearly', data = Age)
plt.xticks(rotation = 45)
plt.ylim(0, 300000)
plt.show()
```

### Visualizing Relationships in Data

**Scatter Plots**

Create a scatter plot of Age and WorkExp.


```python
## Write your code here
AgeWE = """
SELECT Age, WorkExp
FROM main
WHERE Age IS NOT NULL AND WorkExp IS NOT NULL AND Age NOT IN ('Prefer not to say')
GROUP BY Age
ORDER BY Age 
"""
AgeWE = pd.read_sql_query(AgeWE, conn)

plt.figure(figsize=(15,8))
sns.scatterplot(data = AgeWE, x = 'Age', y = 'WorkExp', s=150)
plt.xticks(rotation = 45)
plt.title("Work Experience by Age Group")
plt.ylabel("Years of Experience")
plt.xlabel("Age Group")
plt.tight_layout()
plt.grid()
plt.show()
```

**Bubble Plots**

Create a bubble plot of `TimeSearching` and `Frustration` using the Age column as the bubble size.


```python
## Write your code here
df1 = """
SELECT Age, TimeSearching, Frustration
FROM main
WHERE Age IS NOT NULL AND TimeSearching IS NOT NULL AND Frustration IS NOT NULL AND Age NOT IN ('Prefer not to say')
ORDER BY TimeSearching
"""
df1 = pd.read_sql_query(df1, conn)

df1['FrustList'] = df1['Frustration'].str.split(';').explode('FrustList')
df1['FrustList'] = df1['FrustList'].str.strip()

df1['FrustList'] = df1['FrustList'].str.strip()

plt.figure(figsize=(10, 6))
sns.scatterplot(x = 'TimeSearching', y = 'FrustList', hue = 'Age', s = 500, data = df1)
plt.legend(loc='center left',
    bbox_to_anchor=(1, 0.5),
    title='Age Group'
)
plt.xticks(rotation=45)
plt.show()
```

### Visualizing Composition of Data

**Pie Charts**

Create a pie chart of the top 5 databases(`DatabaseWantToWorkWith`) that respondents wish to learn next year.


```python
## Write your code here
df2 = """
SELECT DatabaseWantToWorkWith
FROM main
WHERE DatabaseWantToWorkWith IS NOT NULL 
"""
df2 = pd.read_sql_query(df2, conn)

df2['DataList'] = df2['DatabaseWantToWorkWith'].str.split(';').explode('DataList')

df2['DataList'] = df2['DataList'].str.strip()

db_counts = df2['DataList'].value_counts().head(5)
plt.pie(db_counts.values, labels = db_counts.index, autopct = '%1.1f%%', startangle = 140)
plt.show()
```

**Stacked Charts** 

Create a stacked bar chart of median `TimeSearching` and `TimeAnswering` for the age group 30 to 35.


```python
## Write your code here
df3 = """
SELECT TimeSearching, TimeAnswering, Age
FROM main
"""
df3 = pd.read_sql_query(df3, conn)

TS_mapping = {
    '30-60 minutes a day' : (30+60)/2,
    'Less than 15 minutes a day' : 15,
    '60-120 minutes a day': (60+120)/2,
    '15-30 minutes a day' : (15+30)/2,
    'Over 120 minutes a day': 120}

TA_mapping = {
    '30-60 minutes a day' : (30+60)/2,
    'Less than 15 minutes a day' : 15,
    '60-120 minutes a day': (60+120)/2,
    '15-30 minutes a day' : (15+30)/2,
    'Over 120 minutes a day': 120}

age_mapping = {
    '25-34 years old': (25+34)/2,
    '35-44 years old': (35+44)/2,
    '18-24 years old': (18+24)/2,
    '45-54 years old': (45+54)/2,
    '55-64 years old': (55+64)/2,
    'Under 18 years old': 18,
    '65 years or older': 65 }

df3['TimeSearching_2'] = df3['TimeSearching'].map(TS_mapping)
df3['TimeSearching_2'].astype('float')

df3['TimeAnswering_2'] = df3['TimeAnswering'].map(TA_mapping)
df3['TimeAnswering_2'].astype('float')

df3['Age'] = df3['Age'].map(age_mapping)
df3['Age'] = df3['Age'].astype('float')

mean_age = df3['Age'].mean(axis = 0)

df3['Age'].fillna(mean_age)

range30_35 = df3[(df3['Age'] >= 30) & (df3['Age'] <= 35)]

med_TS_range30_35 = df3['TimeSearching_2'].median()
med_TA_range30_35 = df3['TimeAnswering_2'].median()

stacked_range30_35 = pd.DataFrame({
    'TimeSearching_Median': [med_TS_range30_35],
    'TimeAnswering_Median': [med_TA_range30_35]
})

fig, ax = plt.subplots(figsize=(10, 6))
stacked_range30_35.plot(kind='bar', stacked=True, ax=ax)
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{int(height)}',
                (p.get_x() + p.get_width() / 2., height),
                ha='center', va='bottom', fontsize=8)
plt.xticks([0], ['Age 30–35'], rotation=0)
plt.xlabel('Time Category')
plt.ylabel('Median Time (minutes)')
plt.title('Median Time Searching and Answering for Age Group 30-35')
plt.xticks(rotation=0) 
plt.show()
```

### Visualizing Comparison of Data

**Line Chart**

Plot the median `CompTotal` for all ages from 45 to 60.


```python
## Write your code here
df4 = """
SELECT CompTotal, Age
FROM main
WHERE CompTotal IS NOT NULL
"""
df4 = pd.read_sql_query(df4, conn)

age_mapping = {
    '25-34 years old': (25+34)/2,
    '35-44 years old': (35+44)/2,
    '18-24 years old': (18+24)/2,
    '45-54 years old': (45+54)/2,
    '55-64 years old': (55+64)/2,
    'Under 18 years old': 18,
    '65 years or older': 65 }

df4['Age'] = df4['Age'].map(age_mapping)
df4['Age'] = df4['Age'].astype('float')

mean_age = df4['Age'].mean(axis = 0)

df4['Age'].fillna(mean_age)

range45_60 = df4[(df4['Age'] >= 45) & (df4['Age'] <= 60)]

median_by_age = range45_60.groupby('Age')['CompTotal'].median().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(median_by_age['Age'], median_by_age['CompTotal'], marker='o')
plt.xlabel('Age')
plt.ylabel('Median Compensation')
plt.title('Median Compensation for Ages 45 to 60')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Bar Chart**

Create a horizontal bar chart using the `MainBranch` column.


```python
## Write your code here
df5 = """
SELECT MainBranch, COUNT(*) AS Total
FROM main
GROUP BY Mainbranch
"""
df5 = pd.read_sql_query(df5, conn)

fig, ax = plt.subplots(figsize=(15, 4))
sns.barplot(data=df5, x='Total', y='MainBranch', orient = 'h')
for p in ax.patches:
    width = p.get_width()
    ax.annotate(f'{int(width)}',
                (width + 200, p.get_y() + p.get_height() / 2),
                ha='left', va='center', fontsize=8, fontweight='bold')
plt.title('Respondents by Main Branch')
plt.xlabel('Total')
plt.ylabel('Main Branch')
plt.tight_layout()
plt.show()
```
